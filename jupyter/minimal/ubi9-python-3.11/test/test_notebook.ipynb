```python
from pathlib import Path
from configparser import ConfigParser
import unittest
import jupyterlab as jp
from platform import python_version
import nbdime
import nbgitpuller

def get_major_minor(s):
    return '.'.join(s.split('.')[:2])

def load_version_overrides_if_present() -> ConfigParser:
    config = ConfigParser()
    config_file = Path('./version_overrides.ini')

    if config_file.is_file():
        config.read(config_file)  

    return config      

class TestPythonVersion(unittest.TestCase):
    def test_version(self):
        expected_major_minor = config.get(CONFIG_SECTION, 'python', fallback = '3.11')
        actual_major_minor = get_major_minor(python_version())
        self.assertEqual(actual_major_minor, expected_major_minor, "incorrect version")

class TestDependenciesVersions(unittest.TestCase):
    def test_jupyter_version(self):
        expected_major_minor = config.get(CONFIG_SECTION, 'jupyterlab', fallback = '4.2')
        actual_major_minor = get_major_minor(jp.__version__)
        self.assertEqual(actual_major_minor, expected_major_minor, "incorrect version")

    def test_nbgitpuller_version(self):
        expected_major_minor = config.get(CONFIG_SECTION, 'nbgitpuller', fallback = '1.2')
        actual_major_minor = get_major_minor(nbgitpuller.__version__)
        self.assertEqual(actual_major_minor, expected_major_minor, "incorrect version")

    def test_nbdime_version(self):
        expected_major_minor = config.get(CONFIG_SECTION, 'nbdime', fallback = '4.0')
        actual_major_minor = get_major_minor(nbdime.__version__)
        self.assertEqual(actual_major_minor, expected_major_minor, "incorrect version")

CONFIG_SECTION = 'minimal'
config = load_version_overrides_if_present()
unittest.main(argv=[''], verbosity=2, exit=False)
```
